# Analysis of Autoregulation Metrics

## Data Loading

```python
import pandas as pd
import numpy as np

# CSV beolvasása
df = pd.read_csv('specific_indices_matrix.csv')
```

## Correlation Analysis

```python


# PatientID oszlop eltávolítása
data = df.drop('PatientID', axis=1)

# Csak azokat a sorokat távolítjuk el, ahol MINDEN érték NaN
data_clean = data.dropna(how='all')

correlation_matrix = data_clean.corr()

# Eredmény kiírása
print("Korrelációs mátrix:")
print(correlation_matrix)

# Opcionálisan mentés CSV-be
correlation_matrix.to_csv('correlation_matrix.csv')
print(f"\nMátrix mérete: {correlation_matrix.shape}")
print(f"Összes beteg száma: {len(data)}")
print(f"Mutatók száma: {len(data.columns)}")

# Statisztika a hiányzó adatokról
missing_info = data.isnull().sum()
print(f"\nHiányzó adatok száma oszloponként (első 5):")
print(missing_info.head())
```
